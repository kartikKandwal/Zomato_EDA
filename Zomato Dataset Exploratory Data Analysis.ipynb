```python
df_country=pd.read_excel('Country-Code.xlsx')
```

```python
df_country.head()
```

```python
# merged df_country to df
final_df=pd.merge(df,df_country,on='Country Code', how='left')
```

```python
final_df.head()
```

```python
# To check Data Types
final_df.dtypes
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
country_name=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
# Pie chart
plt.pie(country_val,labels=country_name)
```

```python
# Pie chart- Top 3
plt.pie(country_val[:3],labels=country_name[:3],autopct='%1.2f%%')
```

Observation: Zomato has 94.3% are from india 

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Counts'})
```

# Observation
1. When Rating is between 4.5 to 4.9---> Excellent
2. When Rating are between 4.0 to 3.4--->very good
3. when Rating is between 3.5 to 3.9----> good
4. when Rating is between 3.0 to 3.4----> average
5. when Rating is between 2.5 to 2.9----> average
6. when Rating is between 2.0 to 2.4----> average

```python
ratings
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(data=ratings,x="Aggregate rating", y="Rating Counts" )
```

```python
sns.barplot(data=ratings,x="Aggregate rating", y="Rating Counts",hue='Rating color', palette=['white','red','orange','yellow','green','green'] )
```

Observation:
1. Not Rated count is very high
2. Maximum number of rating are between 2.5 to 3.4

```python
sns.countplot(data=ratings,x="Rating color", palette=['black','red','orange','yellow','green','green'] )
```
